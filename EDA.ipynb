# Data Cleaning and Preparation:

```python
df=pd.read_csv("Cardiotocographic.csv")
df
```

```python
df.isnull().sum()
```

```python
df.fillna(df.median(),inplace=True)
```

```python
df
```

```python
df['LB']=pd.to_numeric(df['LB'],errors='coerce')  # To remove inconsistency for one column
```

```python
df=df.apply(pd.to_numeric,errors='coerce')   # To remmove inconsistencies from all string to numeric for all columns
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(keep='first',ignore_index=True,inplace=True)   # Removing Duplicates
```

```python
df.duplicated().sum()
```

```python
import seaborn as sns
sns.boxplot(data=df)
plt.show()       # Before Removimg the Outliers
```

```python
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
Q1,Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
l_e=Q1-1.5*IQR
u_e=Q3+1.5*IQR
l_e,u_e
```

```python
df
```

```python
df1=df[abs((df>l_e)|((df<u_e))).all(axis=1)]
```

```python
df1
```

```python
sns.boxplot(df1)             # There are No outliers
```

```python
df1.describe()
```
